```python
import pandas as pd

# Загрузим датасет
df = pd.read_csv('terrorism.csv')

def display_first_100_rows(df):
    if isinstance(df, pd.DataFrame):
        display(df.head(100))
    else:
        print("Переданный объект не является датафреймом Pandas.")

display_first_100_rows(df)
```

### Шаг 2: Обработка пропусков

```python
# Замена пропусков в 'Code'
df['Code'] = df['Code'].fillna('unknown')
display_first_100_rows(df)
```

### Шаг 3: Кодирование категориальных признаков


```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
for column in ['Code']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
display_first_100_rows(df)
```

```python
import matplotlib.pyplot as plt

# Гистограмма для столбца Year после масштабирования
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1) # Первый plot в ряду из двух графиков
plt.hist(df['Year'], bins=20)
plt.title('Распределение Year после масштабирования')
plt.xlabel('Year (стандартизированные значения)')
plt.ylabel('Количество')

# Гистограмма для столбца Terrorist attacks после масштабирования
plt.subplot(1, 2, 2) # Второй plot в ряду из двух графиков
plt.hist(df['Terrorist attacks'], bins=20)
plt.title('Распределение Terrorist attacks после масштабирования')
plt.xlabel('Terrorist attacks (стандартизированные значения)')
plt.ylabel('Количество')

plt.tight_layout()
plt.show()
```

### Шаг 4. Масштабирование данных

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Year', 'Terrorist attacks']] = scaler.fit_transform(df[['Year', 'Terrorist attacks']])
display_first_100_rows(df)
```

```python
import matplotlib.pyplot as plt

# Гистограмма для столбца Year после масштабирования
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1) # Первый plot в ряду из двух графиков
plt.hist(df['Year'], bins=20)
plt.title('Распределение Year после масштабирования')
plt.xlabel('Year (стандартизированные значения)')
plt.ylabel('Количество')

# Гистограмма для столбца Terrorist attacks после масштабирования
plt.subplot(1, 2, 2) # Второй plot в ряду из двух графиков
plt.hist(df['Terrorist attacks'], bins=20)
plt.title('Распределение Terrorist attacks после масштабирования')
plt.xlabel('Terrorist attacks (стандартизированные значения)')
plt.ylabel('Количество')

plt.tight_layout()
plt.show()
```
